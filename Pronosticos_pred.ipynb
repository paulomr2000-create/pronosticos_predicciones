Separacion de caracteristicas que se utilizaran para predecir (independientes) y la que quiero predecir (dependiente)

```python
X = members.drop('Churn', axis=1)
y = members['Churn']

print("Forma de X:", X.shape)
print("Forma de y:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25, 
    random_state=12345,
    stratify=y  # Mantiene la proporción de clases
)

print("Tamaño del conjunto de entrenamiento:", X_train.shape[0])
print("Tamaño del conjunto de validación:", X_test.shape[0])
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Modelo para predecir la cancelacion

```python

lr_model = LogisticRegression(random_state=12345)
lr_model.fit(X_train_scaled, y_train)
lr_predictions = lr_model.predict(X_test_scaled)

# Bosque aleatorio
rf_model = RandomForestClassifier(random_state=12345)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)


print("Regresión Logística:")
print(f"Exactitud: {accuracy_score(y_test, lr_predictions):.3f}")
print(f"Precisión: {precision_score(y_test, lr_predictions):.3f}")
print(f"Recall: {recall_score(y_test, lr_predictions):.3f}")

print("\nBosque Aleatorio:")
print(f"Exactitud: {accuracy_score(y_test, rf_predictions):.3f}")

print(f"Precisión: {precision_score(y_test, rf_predictions):.3f}")
print(f"Recall: {recall_score(y_test, rf_predictions):.3f}")
```

La regresión logística supera al bosque aleatorio en todas las métricas clave:

2.2% más exactitud
Predice correctamente más casos en general
3.7% más precisión
Cuando predice que un cliente cancelará, es más probable que sea correcto
4.9% más recall
Identifica mejor a los clientes que realmente van a cancelar

Creando Clusters

```python
# Dataset para clustering (sin la columna objetivo)
X_clustering = members.drop('Churn', axis=1)
print("Columnas para clustering:", X_clustering.columns.tolist())
print("Forma del dataset:", X_clustering.shape)
```

```python
# Variables con correlación fuerte y moderada
clustering_features = [
    # Correlación FUERTE (>0.35)
    'Lifetime',
    'Avg_class_frequency_current_month', 
    'Age',
    'Contract_period',
    'Month_to_end_contract',
    
    # Correlación MODERADA (0.15-0.35)
    'Avg_class_frequency_total',
    'Avg_additional_charges_total',
    'Group_visits',
    'Partner',
    'Promo_friends'
]

# Crear dataset para clustering
X_clustering = members[clustering_features]
print("Variables seleccionadas para clustering:")
print(X_clustering.columns.tolist())
print(f"Forma del dataset: {X_clustering.shape}")
```

Estandarizando los datos

```python
scaler_clustering = StandardScaler()
X_clustering_scaled = scaler_clustering.fit_transform(X_clustering)
```

```python
print("Medias después de estandarizar:", X_clustering_scaled.mean(axis=0))
print("Desviaciones estándar:", X_clustering_scaled.std(axis=0))
```

```python
linkage_matrix = linkage(X_clustering_scaled, method='ward')
print("Matriz de linkage creada!")
print(f"Forma de la matriz: {linkage_matrix.shape}")
```

Creando Dendograma

```python
plt.figure(figsize=(15, 8))
dendrogram(linkage_matrix, truncate_mode='level', p=10)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Índice de muestra o (tamaño del cluster)')
plt.ylabel('Distancia')
plt.show()
```

```python
# Analizar las últimas fusiones para encontrar el salto más grande
print("Últimas 10 fusiones (distancias):")
print(linkage_matrix[-10:, 2])

# Calcular diferencias entre fusiones consecutivas
distances = linkage_matrix[:, 2]
differences = []
for i in range(len(distances)-1):
    differences.append(distances[i+1] - distances[i])

# Encontrar el salto más grande
max_diff_index = differences.index(max(differences))
optimal_clusters = len(distances) - max_diff_index

print(f"Salto más grande en el índice: {max_diff_index}")
print(f"Número sugerido de clústeres: {optimal_clusters}")
```

```python
from sklearn.cluster import KMeans

# Especificar n=5 clústeres para K-means
n_clusters = 5

# Crear el modelo K-means
kmeans_model = KMeans(
    n_clusters=n_clusters,
    random_state=12345,  # Para resultados reproducibles
    n_init=10           # Número de inicializaciones
)

# Entrenar el modelo y predecir clústeres
kmeans_labels = kmeans_model.fit_predict(X_clustering_scaled)

print(f"K-means completado con {n_clusters} clústeres")
print(f"Distribución de clústeres:")
for i in range(n_clusters):
    count = sum(kmeans_labels == i)
    percentage = (count / len(kmeans_labels)) * 100
    print(f"Clúster {i}: {count} clientes ({percentage:.1f}%)")
```

```python
#Agregar las etiquetas de clúster al DataFrame
members['Cluster'] = kmeans_labels

# Calcular valores medios por clúster
cluster_means = members.groupby('Cluster').mean()
print(cluster_means)

```

Cluster 1 y 4 son los clientes con mas baja cancelacion y cluster 0 y 3 es de mas alto porcentaje de cancenlacion

```python
# Visualizar distribuciones clave
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Contract_period por clúster
sns.boxplot(data=members, x='Cluster', y='Contract_period', ax=axes[0,0])
axes[0,0].set_title('Distribución de Período de Contrato por Clúster')

# Avg_class_frequency_current_month por clúster
sns.boxplot(data=members, x='Cluster', y='Avg_class_frequency_current_month', ax=axes[0,1])
axes[0,1].set_title('Distribución de Frecuencia de Visitas por Clúster')

# Age por clúster
sns.boxplot(data=members, x='Cluster', y='Age', ax=axes[1,0])
axes[1,0].set_title('Distribución de Edad por Clúster')

# Lifetime por clúster
sns.boxplot(data=members, x='Cluster', y='Lifetime', ax=axes[1,1])
axes[1,1].set_title('Distribución de Tiempo de Vida por Clúster')

plt.tight_layout()
plt.show()
```
